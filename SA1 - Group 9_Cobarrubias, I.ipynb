### II. 1.

```python
sample = [2, 3 ,7, 8 ,10]


def moment0(series, r):
    n = len(series)
    return sum([x**r for x in series]) / n

moments0 = [moment0(sample, r) for r in range(1, 5)]

moments_df_zero = pd.DataFrame(moments0, index=["1st Moment", "2nd Moment", "3rd Moment", "4th Moment"])
print(moments_df_zero)
```

### II. 2.

```python
def moment_mean (series, r):
    n = len(series)
    return sum([(x - np.mean(series))**r for x in series]) / n

moments_mean = [moment_mean(sample, r) for r in range (1, 5)]

moments_df_mean = pd.DataFrame(moments_mean, index=["1st Moment", "2nd Moment", "3rd Moment", "4th Moment"])
print(moments_df_mean)
```

### III

Given:
$ m'_4 = m_4 + 4h m_3 + 6h^2 m_2 + h^4, \quad \text{where } h = m'_1 = \mu. $



$ m'_r = E[x^r]. $

We can express $ x $ as $ (x - \mu) + \mu $, where $ \mu = E[x] = m'_1. $

Expanding using the binomial theorem:  
$$
(X - \mu + \mu)^r = \sum_{k=0}^{r} \binom{r}{k} \mu^{r-k}(X - \mu)^k
$$

Taking expectations of both sides:  
$$
m'_r = \sum_{k=0}^{r} \binom{r}{k} \mu^{r-k} E[(X - \mu)^k]
$$

By definition, $ E[(x - \mu)^k] = m_k $, the $k$th central moment.  
Hence,  
$$
m'_r = \sum_{k=0}^{r} \binom{r}{k} \mu^{r-k} m_k
$$

Now, for $ r = 4 $:  
$$
\begin{aligned}
m'_4 &= \binom{4}{0}\mu^4 m_0 + \binom{4}{1}\mu^3 m_1 + \binom{4}{2}\mu^2 m_2 + \binom{4}{3}\mu m_3 + \binom{4}{4}m_4 \\
&= (1)\mu^4(1) + 4\mu^3(0) + 6\mu^2 m_2 + 4\mu m_3 + m_4 \\
&= m_4 + 4\mu m_3 + 6\mu^2 m_2 + \mu^4
\end{aligned}
$$

Since $ h = \mu = m'_1 $, it follows that  
$$
m'_4 = m_4 + 4h m_3 + 6h^2 m_2 + h^4
$$

∎
